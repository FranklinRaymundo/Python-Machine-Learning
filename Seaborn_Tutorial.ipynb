## Seaborn

Seaborn es un paquete complementario a matplotlib, el cual provee de una interfaz de alto nivel para plotear gráficos estadísticos. Seaborn extiende matplotlib. "Si matplotlib intenta hacer fácil las cosas fáciles y posible las cosas difíciles", entonces Seaborn intenta hacer fácil un conjunto de cosas difíciles. Una de las cosas difíciles son los parámetros por default de matplotlib. Seaborn tiene sus propios parámetros.

```python
# Load Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Draw a violin plot
sns.violinplot(x='total_bill',data=tips)
plt.show()
```

Entre algunas de las otras interrogantes están, ¿Cómo traer los colores de seaborn en matplotlib plots. Podemos utilizar la función color_pallete() para definir un mapa de colores y el número de colores con el argumento n_colors. En este caso el ejemplo, asumirá que se tienen 5 labels asignadas a data points data1 y data2.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap

import numpy as np

# Definir una variable N
N = 500

# Construir el colormap
current_pallete = sns.color_palette('muted', n_colors=5)
cmap = ListedColormap(sns.color_palette(current_pallete).as_hex())

# Inicializar la data
data1 = np.random.randn(N)
data2 = np.random.randn(N)

# Asumir que hay 5 labels posibles.
colors = np.random.randint(0,5,N)

# Crear un scatterplot
plt.scatter(data1,data2,c = colors, cmap=cmap)

# Añadir una barra de colores
plt.colorbar()

# Mostrar el plot
plt.show()
```

## Pokemon DataSet 

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
# Loading dataset
import pandas as pd
import numpy as np

pokemon = pd.read_csv('Pokemon.csv',index_col=0,encoding='cp1252')
```

```python
pokemon.head()
```

Una de las fortalezas de seaborn es su variedad de funciones de ploteo.  
Podemos plotear un scatter plot de la siguiente manera:

```python
sns.lmplot(x='Attack',y='Defense',data=pokemon)
plt.show()
```

saeaborn de por si no tiene una función para un scatter plot, de hecho la función lmplot ajusta y plotea una regresión lineal. Pero esta función al igual que muchas incluidas en este paquete contienes argumentos que pueden modificarse, haciendolo podemos dibujar un scatterplot.

```python
sns.lmplot(x='Attack',y='Defense',data=pokemon,fit_reg=False,hue='Stage')
```

Notar que hemos desactivado la función fit_reg y hemos empleado el atributo hue. El cual es útil para expresar una tercera dimensión de información usando el atributo color.  
Vamos a customizar las gráficas usando MatplotLib. Recordar que Seaborn es un interfaz de alto nivel de Matplotlib. De nuestra experiencia, seaborn resulta muy útil, pero a veces será necesario MatplotLib, por ejemplot si deseamos customizar los ejes.

```python
sns.lmplot(x='Attack',y='Defense',data=pokemon,fit_reg=False,hue='Stage')
plt.xlim(0,None)
plt.ylim(0,None)
```

Pandas juega un papel muy importante en las gráficas con seaborn. Las funciones de seaborn se benefician de un dataframe razonablemente formateado.

```python
# Ploteo de Boxplot
sns.boxplot(data=pokemon)
```

Está bien para empezar, pero hay unas columnas que quisieramos remover. Por ejemplo, el Total porque estamos viendo estadísticos individuales, Stage y Legendary porque no son estadísticos de combate.

```python
stats_pokemon = pokemon.drop(['Total','Stage','Legendary'],axis=1)
sns.boxplot(data=stats_pokemon)
```

Otra ventaja de seaborn es que viene con estilos de temas decentes. Siendo 'darkgrid' el que está por defecto.  
Plotearemos una gráfica violín, el cual es una alternativa a los boxplots, muestran la distribución(a través del ancho del violín) en lugar de solo estadísticos resumen.

```python
sns.set_style('whitegrid')
sns.violinplot(x='Type 1',y='Attack',data=pokemon)
```

Por ejemplo, podemos notar que los pokemones de tipo Dragons poseen valores de ataque mayores a los de tipo Ghost, pero también tienen mayor varianza.  
Los fanáticos del pokemon, pueden no estar de acuerdo con la gráfica. Los de tipo Grass de rosado?, los de agua de color anaranjado? Arreglemos esto..

Afortunadamente seaborn permite customizar las paletas de colores. Podemos simplemente crear una lista ordenada de valores hex de colores deseados.

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
sns.violinplot(x='Type 1',y='Attack',data=pokemon,palette=pkmn_type_colors)
```

Los violin plots son buenos para visualizar distribución. Sin embargo, dado que tenemos solo 151 pokemons en el dataset, lo conveniente sería mostrar cada uno de los puntos. Para lograr esto, utilizamos los swarm plots. Esta gráfica mostrará cada punto apilando aquellos con valores similares.

```python
sns.swarmplot(x='Type 1',y='Attack',data=pokemon,palette=pkmn_type_colors)
```

Esto es útil. Pero.¿Podemos combinar nuestro swarm plot y violin plot? Después de todo muestran la misma información. La respuesta es SI.  
De hecho es sencillo sobreponer gráficas usando seaborn. Sigamos los siguientes pasos:  
1. Haremos nuestra figura más grande usando Matplotlib.
2. Plotearemos el gráfico violin. Usaremos el parámetro inner=None para remover las barras dentro del de los violines.
3. Luego plotearemos el swarm plot. Esta vez mostraremos los puntos de color negro para remarcar más.
4. Luego, establecemos un título usando MatplotLib.

```python
plt.figure(figsize=(10,6))

# Creamos el gráfico violin
sns.violinplot(x='Type 1',y='Attack',palette=pkmn_type_colors,data=pokemon,inner=None)

# Crearemos el swarm plot
sns.swarmplot(x='Type 1',y='Attack',data=pokemon,color='k',alpha=0.7 # hace la gráfica ligeramente transparente
             )

plt.title('Ataque por tipo')
```

Podemos repetir la gráfica para cada estadístico. Pero también podemos combinar la información en un solo chart. Antes, tenemos que transformar la estructura de la data. Notamos del dataframe inicial que cada estadístico está en una columna separada. Lo que buscaremos es unirlas en una sola columna. Para esto utilizaremos la función melt() y realizaremos los siguientes pasos:
1. Seleccionar el dataframe a fundir
2. Seleccionaremos las variables ID a permanecer (Pandas fundirá las otras variables)
3. Determinar un nuevo nombre para la variable fundida.

```python
# Melt dataframe
stats_pokemon.head()
```

```python
melted_pokemon = pd.melt(stats_pokemon,id_vars=['Name','Type 1','Type 2'],var_name='Stat')
melted_pokemon.head()
```

Las 6 columnas estdísticas fueron fundidos a una sola, y la nueva columna Stat indica estas 6 variables. En este sentido, el nuevo dataframe debería tener 6 veces la cantidad de filas que el original.

```python
print(stats_pokemon.shape)
print(melted_pokemon.shape)
```

```python
# Swarm plot
plt.figure(figsize=(10,6))
sns.swarmplot(x='Stat',y='value',data=melted_pokemon,hue='Type 1',
             split=True,# Separar puntos por hue
             palette = pkmn_type_colors)

plt.ylim(0,260)

# Mostrar la leyenda a la derecha
plt.legend(bbox_to_anchor = (1,1), loc=2)
```
